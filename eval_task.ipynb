<h3>Importing Libraries</h3>

```python
import pandas as pd # for file reading and preprocessing
import matplotlib.pyplot as plt # for visualization
import seaborn as sns # for visualization
from sklearn.preprocessing import LabelEncoder # for label encoding
```

<h3 style="color:lightblue">Reading the file and getting information</h3>

```python
file=pd.read_csv('Titanic-Dataset.csv')
file.info()
```

<pre>
From above information we get to know that the dataset has total 12 columns out of which 2 are of float64 type ,5 are of int64 type and 5 are of object type.
This dataset contains information about the passengers who were boarded on the ship that got wrecked , containing information about their name , sex , age etc.
The columns 'age' , 'cabin' and 'embarked' consists null values.
Total size of dataset is 83.7+ KB</pre>

<h3 style="color:lightgreen">Handling Null Values</h3>

```python
file['Age']=file['Age'].fillna(file['Age'].mean()) # filling null values in 'Age' column with its mean
file['Cabin']=file['Cabin'].fillna('Unknown') # filling null values with 'Unknown'
file['Embarked']=file['Embarked'].fillna('NA') # filling null values with the mode
file.info()
```

<h3 style="color:indigo">Correlation Matrix</h3>

```python
corr=file.select_dtypes(include=['int64','float64']).drop('PassengerId',axis=1)
cm=corr.corr()
plt.figure(figsize=(8,6))
sns.heatmap(cm,annot=True,fmt='.2f',cmap='Blues',linewidths=0.5)
plt.show
```

<h3 style="color:green">Number of Survived</h3>

```python
surv=file['Survived'].value_counts()
surv
```

```python
colors=['red','blue']
plt.figure(figsize=(15,6))
surv.plot(kind='bar',color=colors)
plt.xlabel('Passengers')
plt.ylabel('Number')
plt.title('Ratio of Survived to Deceased')
plt.grid(True)
plt.show()
```

<pre>As we can see from the above graphs the 'red' bar shows the number of passengers who died and 'blue' bar shows the number of passengers who survived</pre>

<h3 style="color:orange">Count of passengers by 'Sex'</h3>

```python
print(file['Sex'].value_counts())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(y='Sex',data=file)
plt.title("Count of Passengers by Sex")
plt.show()
```

<pre>There were a total of 891 passengers boarded on the ship out of which 577 were male and 314 were females</pre>
